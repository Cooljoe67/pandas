## 1.&nbsp;  Plotting 1 variable

### 1.1.&nbsp;  `.hist()`

Histograms allow you to view the distribution of continuous numerical information. They are great for exploring data.

```python
df['Salary'].sort_values()
```

```python
# histogram of the salary column
df['Salary'].hist();
```

From the histogram above we can see that salaries cluster into two groups: those with salaries roughly below 60,000€ and those with salaries around 70,000€ and above.

Pandas plotting is limited in terms of tweaking/customizing the plot, but you can still change some basic stuff. Here we'll adjust the number of bins in the histogram and the size of the figure.

```python
# bins are the amount of coulmns the histogram is divided into. The more bins, the finer the detail that can be seen
# try changing the number of bins and running this cell again, how has the histogram changed?

# why not try also changing the figsize numbers one at a time and running the cell again, what changed?

df['Salary'].hist(bins=12, figsize=(12,6));
```

### 1.2.&nbsp; `.plot.bar()`

```python
#df.plot.bar()
df['Salary'].plot(kind='line')
```

Bar plots are for visualising categorical data, such as the number of people working in each department. This would require two steps:
1. Group the categories and generate the frequencies with `value_counts()`
2. Use `.plot.bar()` to create a graph

For simple frequency plots, which show us how many values are in each category:

1. You should first group the categories and generate the frequencies with `value_counts()`

2. You should use a **bar plot** rather than a histogram, as we're now using categorical data instead of numerical
  - Numerical data - histogram
  - Categorical data - bar plot

Let's see this in action:

```python
# 1. group the categories and generate the frequencies with value_counts()
colour_freq = df['Favourite_Colour'].value_counts()
colour_freq
```

```python
# 2. plot our value_counts from above as a bar plot
colour_freq.plot.bar();
```

Similar to `.hist()`, there are some parameters available in `.plot()` that you can use to improve your plots, such as color or size. The documentation is always the ultimate source of truth when it comes to understanding which parameters you can tweak and how to do it:

- [These are the docs specifically for `DataFrame.plot.bar()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.bar.html)
- [Here are the docs for `DataFrame.plot()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html#pandas.DataFrame.plot), with arguments that apply to all types of plots, not just bar plots.

```python
# changing the colors of the bars
colour_freq.plot.bar(color=["purple", "blue", "red", "yellow", "green", "pink", "orange"]);
```

## 2.&nbsp; Plotting 2 or more variables

### 2.1.&nbsp; `.plot.scatter()`

Whenever you want to see if there is a relationship between two numerical variables, you could compute something complicated like a linear correlation coefficient, or perform more sophisticated statistical tests. For a nice intuitive grasp of the numbers, however, nothing beats a good old scatter plot.

```python
df.plot.scatter(x='Age', y='Salary');
```

Scatter plots allow us to quickly see a strong/weak correlation at a glance. You can see a slight trend in the above scatter plot, showing that, on the whole, older people tend to have higher salaries than younger ones.

Let's tweak some of the scatter plots parameters:

- The `c` argument allows **colors** to depend on the values of another (numerical) column.
- The `cmap` argument allows the usage of "**color maps**" to fill the dots. See all available cmaps [here](https://matplotlib.org/stable/tutorials/colors/colormaps.html)
- The `s` argument modifies the **size of the dots**.
- The `figsize` argument defines the **size of the whole plot**.



  > **Note:** Some of these options might only be available if you have one of the latest Pandas versions, as `pandas.plot()` is one of the areas where pandas is growing. You can check your pandas version with `pd.__version__` in the notebook. If your version is substantially older than the one shown in [here](https://pandas.pydata.org/docs/index.html), consider updating pandas by typing on your Terminal `pip install --upgrade pandas`.



```python
# This is the same scatter plot as above, just bigger, more colorful, and now with an added third variable 'Rating'.
df.plot.scatter(x='Age',
                y='Salary',
                c='Rating',
                s=350,
                cmap="viridis",
                figsize=(12,8)
                );
```
