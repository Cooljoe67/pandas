#### Exercise 1
How many sets were released in each year?

```python
sets_df.groupby("year")['set_num'].count().sort_values(ascending=False)
```

#### Exercise 2
In which five years were the most sets released?

```python
sets_df.groupby("year")['set_num'].count().nlargest(5)
```

#### Exercise 3
What is the median number of parts and the number of sets in each theme?

```python
sets_df.groupby("theme_id")['num_parts'].agg(Median_num_parts=('median'), cnt_sets=('count'))
```

#### Exercise 4

```python
sets_df.groupby(["theme_id","year"])[['year','theme_id']].agg('count').nlargest(1,"theme_id")
```

#### Exercise 5
What is the name of the theme with the most sets associated?

```python
sets_df.merge(themes_df, left_on='theme_id', right_on='id').groupby(['theme_id','name_y'])[["theme_id","name_y"]].size().reset_index().nlargest(1,0)
```

#### Exercise 6
What minifigs are in the set 'Traitor Transport (with Cave)'?    

> If two DataFrames have a column of the same name (other than the merging column), you can control the columns' names in the output by using the parameter `suffixes` in the merge.

```python
sets_df.loc[sets_df["name"] == 'Traitor Transport (with Cave)'].merge(inventories_df,left_on="set_num",right_on='set_num',suffixes=('_sets','_inv')).merge(inventory_minifigs_df,left_on='id_inv',right_on='inventory_id')
```
